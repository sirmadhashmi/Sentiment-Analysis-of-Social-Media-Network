```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
#X = vect.fit_transform([y[1:-1] for y in [x.strip() for x in uSA_df['words'][0][1:].split(',')]]) 
X = vect.fit_transform(uSA_df['words'].astype(str))
y = uSA_df['Sentiment']
```

```python
print(len(vect.vocabulary_))
print(vect.vocabulary_)
vect.get_feature_names()
idf = vect.idf_
#print(dict(zip(vect.get_feature_names(), idf)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
classifier = ExtraTreesClassifier(n_estimators=1000, verbose=2, n_jobs= 2, random_state=0)  
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred)) 
print(accuracy_score(y_test, y_pred))  
```

```python
import matplotlib.pyplot as plt
import itertools
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
plot_confusion_matrix(confusion_matrix(y_test, y_pred1), classes=['-1','0','1'], cmap=plt.cm.Purples, 
                      title='Confusion matrix ExtraTreeClassifier, without normalization')
```
